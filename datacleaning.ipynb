```python
# removing outliers from the dataset
Q1 = df_cleaned['Price'].quantile(0.25) # first quartile
Q3 = df_cleaned['Price'].quantile(0.75) # third quartile
IQR = Q3 - Q1 # interquartile range
lower_bound = Q1 - 1.5 * IQR # lower bound for outliers
upper_bound = Q3 + 1.5 * IQR # upper bound for outliers
df_no_outliers = df_cleaned[(df_cleaned['Price'] >= lower_bound) & (df_cleaned['Price'] <= upper_bound)] # removing outliers
df_no_outliers
```

```python
# saving the cleaned dataset without outliers
df_no_outliers.to_csv("LAPTOP_PRICE_NO_OUTLIERS.csv", index=False) # saving the cleaned dataset without outliers
```

```python
# reading the cleaned dataset without outliers
df_cleaned = pd.read_csv("LAPTOP_PRICE_NO_OUTLIERS.csv") # reading the cleaned dataset without outliers
df_cleaned
```

```python
# dividing the data into numerical and categorical features
characters= df_cleaned.select_dtypes(include='O') # categorical features
characters.head()
```

```python
numericals = df_cleaned.select_dtypes(exclude='O') # numerical features
numericals.head()
```

```python
# removing unnecessary columns from numerical features
numericals = numericals.drop(["Number of Ratings", "Number of Reviews"], axis=1)
numericals.head()
```

```python
characters.head()
```

```python
characters["brand"].value_counts()
```

```python
# encoding categorical features
dt1 = {i : key for key, i in enumerate(list(characters["brand"].unique()))}
dt1
```

```python
# encoding the categorical features using the mapping dictionary
characters["brand"] = characters["brand"].map(dt1)
characters.head()
```

```python
dt2 = {i : key for key, i in enumerate(list(characters["processor_brand"].unique()))}
dt2
```

```python
characters["processor_brand"] = characters["processor_brand"].map(dt2)
characters.head()
```

```python
characters = characters.drop(["weight", "warranty", "msoffice"], axis=1)
characters.head()
```

```python
# encoding the remaining categorical features
dt3 = {i : key for key, i in enumerate(list(characters["processor_name"].unique()))}
dt4 = {i : key for key, i in enumerate(list(characters["processor_gnrtn"].unique()))}
dt5 = {i : key for key, i in enumerate(list(characters["ram_gb"].unique()))}
dt6 = {i : key for key, i in enumerate(list(characters["ram_type"].unique()))}
dt7 = {i : key for key, i in enumerate(list(characters["ssd"].unique()))}
dt8 = {i : key for key, i in enumerate(list(characters["hdd"].unique()))}
dt9 = {i : key for key, i in enumerate(list(characters["os"].unique()))}
dt10= {i : key for key, i in enumerate(list(characters["os_bit"].unique()))}
dt11= {i : key for key, i in enumerate(list(characters["graphic_card_gb"].unique()))}
dt12= {i : key for key, i in enumerate(list(characters["Touchscreen"].unique()))}
dt13= {i : key for key, i in enumerate(list(characters["rating"].unique()))}
```

```python
# encoding the remaining categorical features using the mapping dictionary
characters["processor_name"] = characters["processor_name"].map(dt3)
characters["processor_gnrtn"] = characters["processor_gnrtn"].map(dt4)
characters["ram_gb"] = characters["ram_gb"].map(dt5)
characters["ram_type"] = characters["ram_type"].map(dt6)
characters["ssd"] = characters["ssd"].map(dt7)
characters["hdd"] = characters["hdd"].map(dt8)
characters["os"] = characters["os"].map(dt9)
characters["os_bit"] = characters["os_bit"].map(dt10)
characters["graphic_card_gb"] = characters["graphic_card_gb"].map(dt11)
characters["Touchscreen"] = characters["Touchscreen"].map(dt12)
characters["rating"] = characters["rating"].map(dt13)
characters.head()
```

```python
# combining the numerical and categorical features
df_final = pd.concat([numericals, characters], axis=1) # combining the numerical and categorical features
df_final
```

```python
# saving the final dataset
df_final.to_csv("LAPTOP_PRICE_FINAL_NUMERIC.csv", index=False) # saving the final dataset
```

```python
# reading the final dataset
df_final = pd.read_csv("LAPTOP_PRICE_FINAL_NUMERIC.csv") # reading the final dataset
```

```python
# getting a correlation matrix
correlation = df_final.corr() # getting the correlation matrix
correlation
```

```python
# plotting the correlation matrix
plt.figure(figsize=(20, 20)) # setting the figure size
sns.heatmap(correlation, annot=True) # plotting the heatmap
plt.title("Correlation Matrix") # setting the title of the plot
```

```python
# dividing the data into features and target variable
y = df_final["Price"] # target variable
x = df_final.drop(["Price"], axis=1) # features
```

```python
# splitting the data into training and testing sets
from sklearn.model_selection import train_test_split # importing the train_test_split function
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42) # splitting the data into training and testing sets
```

```python
# applying linear regression model
from sklearn.linear_model import LinearRegression # importing the LinearRegression class
LR = LinearRegression() # creating an instance of the LinearRegression class
LR.fit(x_train, y_train) # fitting the model to the training data
```

```python
LR.score(x_train, y_train) # getting the score of the model on the training data
```

```python
LR.score(x_test, y_test) # getting the score of the model on the testing data
```
